```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
phis = np.linspace(-np.pi, np.pi, 10000)

a_r = 2
a_phi = np.pi/2

r_min = 0.5
r_max = 7.5

phi_min = -np.pi
phi_max = np.pi

r_0 = 4
phi_0 = 0

# r-phi coordinates
r_e =  r_0 + a_r*np.cos(phis)
phi_e = phi_0 + a_phi*np.sin(phis)

r_e_hor = r_e
phi_e_hor = phi_0 * np.ones(len(r_e))

r_e_vert = r_0 * np.ones(len(r_e))
phi_e_vert = phi_e


# cartesian coordinates
x_e = r_e * np.cos(phi_e)
y_e = r_e * np.sin(phi_e)

x_e_vert = r_e_vert * np.cos(phi_e_vert)
y_e_vert = r_e_vert * np.sin(phi_e_vert)

x_e_hor = r_e_hor * np.cos(phi_e_hor)
y_e_hor = r_e_hor * np.sin(phi_e_hor)
```

```python
border_color = "grey"
border_alpha = 1
border_ls = "-"

semiaxis_color = "silver"
semiaxis_alpha = 0.6
semiaxis_ls = "-"
```

```python
# fix, ax = plt.subplots()
fig, axes = plt.subplots(1,3, dpi=150, figsize=(9,3))

axes = [axes[n] for n in [2,1,0]]

####################
# cartesian
####################
ax = axes[0]

ax.plot(x_e, y_e, solid_capstyle='round')
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")


ax.plot(r_min*np.cos(phis), r_min*np.sin(phis), color = border_color, alpha=border_alpha, ls=border_ls)
ax.plot(r_max*np.cos(phis), r_max*np.sin(phis), color = border_color, alpha=border_alpha, ls=border_ls)
ax.plot([-r_max, -r_min], [0,0], color = border_color, alpha=border_alpha, ls=border_ls)

ax.plot(x_e_vert, y_e_vert, color = semiaxis_color, alpha=semiaxis_alpha, ls=semiaxis_ls, zorder=-10000)
ax.plot(x_e_hor, y_e_hor, color = semiaxis_color, alpha=semiaxis_alpha, ls=semiaxis_ls, zorder=-10000)

ax.set_aspect("equal")

# ticks
ax.set_xticks([-r_max, 0, r_max])
ax.set_xticklabels([r"$-r_\max$", r"$0$", r"$r_\max$"])
ax.set_yticks([-r_max, 0, r_max])
ax.set_yticklabels([r"$-r_\max$", r"$0$", r"$r_\max$"])

####################
# r - arclength
####################

# fix, ax = plt.subplots()
ax = axes[1]

squish_factor = 6
phis_squish = phis/squish_factor

r_e_squish =  r_0 + a_r*np.cos(phis)
phi_e_squish = (phi_0 + a_phi*np.sin(phis))/squish_factor

x_e_squish = r_e_squish * np.cos(phi_e_squish)
y_e_squish = r_e_squish * np.sin(phi_e_squish)

ax.plot(x_e_squish, y_e_squish, solid_capstyle='round')
ax.set_xlabel(r"")
ax.set_ylabel(r"")


ax.plot(r_min*np.cos(phis_squish), r_min*np.sin(phis_squish), color = border_color, alpha=border_alpha, ls=border_ls, solid_capstyle='round')
ax.plot(r_max*np.cos(phis_squish), r_max*np.sin(phis_squish), color = border_color, alpha=border_alpha, ls=border_ls, solid_capstyle='round')

sma = np.sin(phis_squish[-1])
cma = np.cos(phis_squish[-1])

smi = np.sin(phis_squish[0])
cmi = np.cos(phis_squish[0])

ax.plot([r_min*cma, r_max*cma], [r_min*sma, r_max*sma], color = border_color, alpha=border_alpha, ls=border_ls)
ax.plot([r_min*cmi, r_max*cmi], [r_min*smi, r_max*smi], color = border_color, alpha=border_alpha, ls=border_ls)


x_e_vert_squish = r_e_vert * np.cos(phi_e_vert/squish_factor)
y_e_vert_squish = r_e_vert * np.sin(phi_e_vert/squish_factor)

x_e_hor_squish = r_e_hor * np.cos(phi_e_hor/squish_factor)
y_e_hor_squish = r_e_hor * np.sin(phi_e_hor/squish_factor)


ax.plot(x_e_vert_squish, y_e_vert_squish, color = semiaxis_color, alpha=semiaxis_alpha, ls=semiaxis_ls, zorder=-10000)
ax.plot(x_e_hor_squish, y_e_hor_squish, color = semiaxis_color, alpha=semiaxis_alpha, ls=semiaxis_ls, zorder=-10000)

# ticks
ax.set_yticks([])
ax.set_xticks([r_min, r_0, r_max])
ax.set_xticklabels([r"$r_\min$", r"$r_0$", r"$r_\max$"])

# labels
ax.text(r_0, r_0*phis_squish[-1], r"$\phi = \pi$", ha="center", va="bottom", rotation=180/squish_factor)
ax.text(r_0, r_0*phis_squish[0], r"$\phi = -\pi$", ha="center", va="top", rotation=-180/squish_factor)

ax.set_box_aspect(1)

####################
# r-phi
####################

# fix, ax = plt.subplots()
ax = axes[2]

ax.plot(r_e, phi_e, solid_capstyle='round')
ax.set_xlabel(r"$r$")
ax.set_ylabel(r"$\phi$")


ax.plot([r_min, r_min], [phi_min, phi_max], color = border_color, alpha=border_alpha, ls=border_ls)
ax.plot([r_max, r_max], [phi_min, phi_max], color = border_color, alpha=border_alpha, ls=border_ls)
ax.plot([r_min, r_max], [phi_min, phi_min], color = border_color, alpha=border_alpha, ls=border_ls)
ax.plot([r_min, r_max], [phi_max, phi_max], color = border_color, alpha=border_alpha, ls=border_ls)


# ellipse axes
ax.plot(r_e_vert, phi_e_vert, color = semiaxis_color, alpha=semiaxis_alpha, ls=semiaxis_ls, zorder=-10000)
ax.plot(r_e_hor, phi_e_hor, color = semiaxis_color, alpha=semiaxis_alpha, ls=semiaxis_ls, zorder=-10000)


# ticks
ax.set_yticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticklabels([r"$-\pi$", r"$-\pi/2$",r"$0$",r"$\pi/2$",r"$\pi$"])
ax.set_xticks([r_min, r_0, r_max])
ax.set_xticklabels([r"$r_\min$", r"$r_0$", r"$r_\max$"])

ax.set_aspect("equal")
ax.set_box_aspect(1)


####################
# Final touches
####################

    
fig.tight_layout()

fig.savefig("geometries.svg")
fig.savefig("geometries.pdf")
```
